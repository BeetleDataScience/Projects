```python
usuario = "root"
contraseña = "Giant1996."
host = "localhost"
base_de_datos = "chicago_db"

# Crear el motor de conexión
engine = create_engine(f"mysql+pymysql://{usuario}:{contraseña}@{host}/{base_de_datos}")
```

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv")
df.to_sql("chicago_school", con=engine, if_exists="append", index=False)

print("Datos insertados exitosamente.")
```

```python
%load_ext sql
conexion = f"mysql+pymysql://{usuario}:{contraseña}@{host}/{base_de_datos}"
%sql {conexion}
```

```python
#Muestre todas las tablas de la database chicago_db
%sql SHOW TABLES;
```

```python
# Obtenga el numero de columnas de la tabla chicago_school al igual que los nombres de su columna
# su tipo de datos y longitudes
%sql DESCRIBE chicago_school;
```

```python
# ¿Cuántas elementary school (ES) hay en el conjunto de datos?
%sql SELECT COUNT(*) FROM chicago_school WHERE `Elementary, Middle, or High School` = 'ES';
```

```python
# ¿Cuál es la safety_score mas alta?
%sql SELECT MAX(SAFETY_SCORE) FROM chicago_school;
```

```python
# ¿Cuáles son los schools con mayor puntuación en safety_score?
%sql SELECT NAME_OF_SCHOOL FROM chicago_school WHERE SAFETY_SCORE = 99.0
```

```python
# ¿Cuáles son los 10 schools con mayor "Average Student Attendance"?
%sql SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE FROM chicago_school ORDER BY AVERAGE_STUDENT_ATTENDANCE DESC LIMIT 10;
```

```python
# Recuperar la lista de 5 schools con la menor asistencia AVERAGE_STUDENT_ATTENDANCE ordenada de forma ascendente 
# en función de la asistencia.

%sql SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE FROM chicago_school ORDER BY AVERAGE_STUDENT_ATTENDANCE ASC LIMIT 5;
```

```python
# Ahora elimine el signo «%» del resultado de la columna AVERAGE_STUDENT_ATTENDANCE.
%sql SELECT NAME_OF_SCHOOL, REPLACE(AVERAGE_STUDENT_ATTENDANCE, '%', '') FROM chicago_school ORDER BY AVERAGE_STUDENT_ATTENDANCE ASC LIMIT 5;
```

```python
# ¿Qué schools tienen una AVERAGE_STUDENT_ATTENDANCE inferior al 70%?
%sql SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE FROM chicago_school WHERE AVERAGE_STUDENT_ATTENDANCE < 70
```

```python
# Obtenga el total College Enrollment para cada Community Area
%sql SELECT Community_Area_Name, SUM(College_Enrollment) AS TOTAL_ENROLLMENT FROM chicago_school GROUP BY Community_Area_Name 
```

```python
# Obtenga las 5 Community Areas con menor numero de total College Enrollment sorted de forma descendente
%sql SELECT Community_Area_Name, SUM(College_Enrollment) AS TOTAL_ENROLLMENT \
   FROM chicago_school \
   GROUP BY Community_Area_Name \
   ORDER BY TOTAL_ENROLLMENT asc \
   LIMIT 5; 
```
