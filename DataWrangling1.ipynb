```python
import pandas as pd
import numpy as np
import matplotlib.pylab as p
```

```python
#Tenemos el link de los datos pero no tienen el nombre de las columnas
file_path="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"

# creamos una lista con los nombres de las columnas
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
# Lo primero que haremos es crear el dataframe usando pandas
df = pd.read_csv(file_path, header = None)
```

```python
# Añadiremos los valores de las columnas
df.columns = headers
```

```python
#Haremos un pequeña validacion del dataframe 
df.head()
```

**observamos como vemos varios datos perdidos (?), lo primeros que haremos es** 
1. Identificar los datos que faltan
2. Tratar los datos que faltan
3. Corregir el formato de los datos

```python
# 1. Identificar los datos que faltan 
##En nuestro conjunto de datos los valores faltantes vienen representados por (?)
###Los sustituiremos por 'NaN' que es el marcador de valores faltante de default de Python 

df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
# Puede usar la funcion .isnull en su dataframe que le arrojara un valor Booleano de true si es NaN
## o False si no es un valor NaN
missing_data = df.isnull()
missing_data.head(5)
```

```python
# Ahora bien, si es cierto que ya podemos distinguir valores NaN del resto, en una dataframe con 
## cientos o miles de filas y columnas es muy poco probable poder saber este dato con exactitud a 
### simple vista, por esto usaremos un bucle FOR que nos permitara identificar cuantos de estos 
#### valores TRUE existen en cada Columna. 

for column in missing_data.columns.values.tolist():  #obtiene una lista de las columnas
    print(column)
    print (missing_data[column].value_counts()) #utilizando la variable missing_data que hayamos anteriormente podemos contar los valores unicos de cada columna 
    print("")
```

**2. Bien, Ya conocemos la totalidad de datos faltantes. ahora ¿Como trataremos estos datos?**
tenemos 2 opciones 

a. Eliminar la fila o la columna que tenga los datos faltantes. Solo se recomienda cuando la mayoria de los datos de la columna es NaN. 

b. sustituir los datos, ya sea por la media, la frecuencia o cualquier otra funciones pero todo esto dependera del tipo de dato

**Para este ejercicio lo haremos de la siguiente manera**

**Reemplazar por media:**

normalized-losses": 41 datos perdidos
stroke": 4 datos ausentes 
bore": 4 datos faltantes
horsepower": Faltan 2 datos
peak-rpm": Faltan 2 datos

**Reemplazar por frecuencia:**

num-of-doors": Faltan 2 datos, sustitúyalos por «cuatro».
Razón: El 84% de las berlinas son de cuatro puertas. Como cuatro puertas es lo más frecuente, es lo más probable que ocurra

**Eliminar toda la fila:**

price": Faltan 4 datos, simplemente elimine toda la fila
Motivo: Quiere predecir el precio. No puede utilizar ninguna entrada de datos sin datos de precio para la predicción; por lo tanto, cualquier fila ahora sin datos de precio no le es útil.



```python
#Calcularemos la media de la columna "normalized-losses" 
avg_normalized_losses = df["normalized-losses"].astype("float").mean(axis=0) #convertirmos a float los valores de la columna y con .mean sacamos la media sobre las filas de la columna por eso (axis=0)
print("Average of normalized-losses:", avg_normalized_losses)
```

```python
#Y ahora reemplazamos los valores faltantes con el valor encontrado en el punto anterior
df["normalized-losses"] = df["normalized-losses"].replace(np.nan, avg_normalized_losses)
```

```python
# la media de la columna "stroke" 
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average of stroke:", avg_stroke)
```

```python
#Y reemplazamos
df["stroke"] = df["stroke"].replace(np.nan, avg_stroke)
```

```python
# la media de la columna "bore" 
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
#Y reemplazamos
df["bore"] = df["bore"].replace(np.nan, avg_bore)
```

```python
# la media de la columna "horsepower" 
avg_horsepower=df['horsepower'].astype('float').mean(axis=0)
print("Average of horsepower:", avg_horsepower)
```

```python
#Y reemplazamos
df["horsepower"] = df["horsepower"].replace(np.nan, avg_horsepower)
```

```python
# la media de la columna "horsepower" 
avg_peak_rpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average of peak-rpm:", avg_peak_rpm)
```

```python
#Y reemplazamos
df["peak-rpm"] = df["peak-rpm"].replace(np.nan, avg_peak_rpm)
```

```python
## Ahora reemplazaremos por frecuencia la variable num-of-doors
## Veamos la frecuencia de los datos de la columna num-of-doors
df['num-of-doors'].value_counts()
```

```python
# podemos observar que 'four' es el que mas se repite, pero tambien podemos encontrarlo con la siguiente funcino 
df['num-of-doors'].value_counts().idxmax()
```

```python
#Ahora que sabemos el valor, reemplemoslo en el marco de datos 
df["num-of-doors"] = df["num-of-doors"].replace(np.nan, "four")
```

```python
#Finalmente eliminemos las filas que tengan valores NaN en la columna Price
df.dropna(subset=["price"], axis=0, inplace=True)
```

```python
# Restablescamos el indice, porque hemos eliminado filas 
df.reset_index(drop=True, inplace=True)
```

```python
# Ahora volvamos a ver nuestra dataframe, ya no tenemos datos NaN
df.head()
```

Como puede ver arriba, algunas columnas no tienen el tipo de datos correcto. Por ejemplo, los valores numéricos «bore» y «stroke» describen los motores, por lo que debería esperarse que fueran del tipo «float» o «int»; sin embargo, se muestran como tipo «objeto». ## Ahora pasaremos a la normalizacion de datos 

La normalización es el proceso de transformación de los datos en un formato común, que permite al investigador realizar una comparación significativa. Ya que en muchas ocaciones los datos son recogidos de diferentes sitios que pueden usar formato diferentes

**Ejemplo**

Transformar mpg a L/100km:

En su conjunto de datos, las columnas de consumo de combustible «city-mpg» y «highway-mpg» están representadas por la unidad mpg (millas por galón). Suponga que está desarrollando una aplicación en un país que acepta el consumo de combustible con el estándar L/100km.


```python
# Convierte mpg a L/100km por operación matemática (235 dividido por mpg)
df['city-L/100km'] = 235/df["city-mpg"]
```

```python
# transformar mpg a L/100km por operación matemática (235 dividido por mpg)
df["highway-mpg"] = 235/df["highway-mpg"]

# renombre la columna de "highway-mpg" to "highway-L/100km"
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
```

```python
df.head()
```

**¿Por qué normalizar?**

La normalización es el proceso de transformar los valores de varias variables en un rango similar. Las normalizaciones típicas incluyen

escalar la variable para que la media de la variable sea 0
escalar la variable para que la varianza sea 1
escalar la variable para que sus valores oscilen entre 0 y 1
**Ejemplo**

Para demostrar la normalización, supongamos que desea escalar las columnas "length", "width" and "height".

Objetivo: normalizar esas variables para que su valor oscile entre 0 y 1

Método: sustituir el valor original por (valor original)/(valor máximo)

```python
# Podemos hacerlo de la siguiente manera 
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

**Binning**
¿Por qué el binning?

El binning es un proceso de transformación de variables numéricas continuas en «intervalos» categóricos discretos para el análisis agrupado.

**Ejemplo:**

En su conjunto de datos, "horsepower" es una variable de valor real que oscila entre 48 y 288 y tiene 59 valores únicos. ¿Qué ocurre si sólo le interesa la diferencia de precio entre los coches con mucha potencia, potencia media y poca potencia (3 tipos)? Puedes reordenarlos en tres 'bins' para simplificar el análisis.

Utilice el método 'cut' de Pandas para segmentar la columna 'caballos' en 3 grupos.

```python
df["horsepower"]=df["horsepower"].astype(int)
```

```python
#Construimos los bins donde el valor min es el valor minimo de la variable "horsepower" y
## el valor maximo es el valor maximo de la variables "horsepower
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
# creamos los nombres de los set 
group_names = ['Low', 'Medium', 'High']
```

```python
# Y usamos la funcion cut para determinar los valores y lo revisamos 
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
# podemos contar el numero de datos en cada rango 
df["horsepower-binned"].value_counts()
```

```python
# grafiquemos esto que acabemos de hacer 

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

**Variable indicadora**
**¿Qué es una variable indicadora?**

Una variable indicadora (o variable ficticia) es una variable numérica utilizada para etiquetar categorías. Se denominan «ficticias» porque los números en sí no tienen un significado inherente.

**¿Por qué utilizar variables indicadoras**

Se utilizan variables indicadoras para poder utilizar variables categóricas para el análisis de regresión en los módulos posteriores.

**Ejemplo**
La columna "type fuel" tiene dos valores únicos: "gas" o “diesel”. La regresión no entiende de palabras, sólo de números. Para utilizar este atributo en el análisis de regresión, puedes convertir «tipo-combustible» en variables indicadoras.

Utilice el método de Panda 'get_dummies' para asignar valores numéricos a las diferentes categorías de tipo de combustible.

```python
# creemos nuestro dummie 
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
#cambiemos los nombres de las columnas del dummie
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# fusionamos el marco de datos con el dummie" 
df = pd.concat([df, dummy_variable_1], axis=1)
```
